```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Drop rows with missing values in 'new_tests' or 'new_cases'
regression_df = africa_df[['new_tests', 'new_cases']].dropna()

# Prepare the data
X = regression_df[['new_tests']].values
y = regression_df['new_cases'].values

# Train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict for the regression line
X_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_line = model.predict(X_line)

# Plot the scatter plot and regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Data Points')
plt.plot(X_line, y_line, color='red', label='Regression Line')
plt.title("New Tests vs. New Cases in Africa")
plt.xlabel("New Tests")
plt.ylabel("New Cases")
plt.legend()
plt.grid(True)
plt.show()

#- **Outcome**:  
  #- Trains a linear regression model to predict `new_cases` based on `new_tests`.  
  #- Displays a scatter plot of actual data points (`new_tests` vs. `new_cases`).  
  #- Overlays a regression line showing the predicted relationship.  
  #- Helps visualize the correlation between testing and case numbers in Africa.
```

```python
# Line plot of total cases over time
africa_df['date'] = pd.to_datetime(africa_df['date'])
cases_over_time = africa_df.groupby('date')['total_cases'].sum()
cases_over_time.plot(kind='line', figsize=(10, 6), title="Total Cases Over Time in Africa")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.grid(True)
plt.show()

#- **Outcome**:
    #- Displays a line plot showing the trend of total COVID-19 cases in Africa over time.
    #- total cases increased over time, indicating the progression of the pandemic.
    #- with time, casess become consistent. and can be explained due to the vaccination process.
```

```python
# Line plot of total deaths over time
deaths_over_time = africa_df.groupby('date')['total_deaths'].sum()
deaths_over_time.plot(kind='line', figsize=(10, 6), color='red', title="Total Deaths Over Time in Africa")
plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.grid(True)
plt.show()

#- **Outcome**:
    #- Displays a line plot showing the trend of total COVID-19 deaths in Africa over time.
    #- total deaths increased over time, indicating the progression of the pandemic.
    #- with time, deaths become consistent. and can be explained due to the vaccination process.
```

```python
# Bar plot of positivity rates
africa_df['positivity_rate'] = africa_df['new_cases'] / africa_df['new_tests']
positivity_by_country = africa_df.groupby('location')['positivity_rate'].mean().dropna().sort_values()
positivity_by_country.plot(kind='bar', figsize=(12, 6), title="Testing Positivity Rate by Country in Africa")
plt.xlabel("Country")
plt.ylabel("Positivity Rate")
plt.show()
```

```python
# Bar plot for total cases by country
cases_by_country = africa_df.groupby('location')['total_cases'].sum().sort_values()
cases_by_country.plot(kind='barh', figsize=(10, 10), title="Total Cases by Country in Africa", color='green')
plt.xlabel("Total Cases")
plt.ylabel("Country")
plt.show()

#- **Outcome**:
    #- South Africa reported the highest total cases in africa and Keny my country being the 9th in line.
    #- western sahara reported the least n. of cases.
```
